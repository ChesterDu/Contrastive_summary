```python
import torch
logs_with = torch.load("finetune_log.pkl")
logs_without = torch.load("finetune_log_no_pretrain.pkl")
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
steps = [item["step"] for item in logs_with]
acc1 = [item["acc"] for item in logs_with]
acc2 = [item['acc'] for item in logs_without]



plt.plot(steps, acc1, label = 'with contrastive training')
plt.plot(steps, acc2, label = 'without contrastive trainig')
plt.xlabel('Steps')
plt.ylabel("Accuracy")
plt.title("Test Accuracy vs Train Steps")
plt.legend()
```

```python
loss1 = [item["loss"] for item in logs_with]
loss2 = [item["loss"] for item in logs_without]

plt.plot(steps, loss1, label = 'with contrastive training')
plt.plot(steps, loss2, label = 'without contrastive trainig')
plt.xlabel('Steps')
plt.ylabel("loss")
plt.title("Train Loss vs Train Steps")
plt.legend()
```
